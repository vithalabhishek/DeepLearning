```
import numpy as np
```

```
inputs = np.array([[0,1,0],[1,0,1],[1,1,0],[0,0,1]])
outputs = np.array([[0],[1],[1],[0]])
```

```
inputs
```

```
outputs
```

```
np.random.seed(10)
weights = np.random.random((3,1))
weights
```

```
sum = np.dot(inputs,weights) + 0.02
sum
```

```
### Activation Function (sigmoid)
def activation(x):
  return 1/(1+np.exp(-x))
```

```
### Activation Function (sigmoid)
def step_function(x):
  if x<0:
    return 0
  else:
    return 1
```

```
ypred = activation(sum)
ypred
```

```
ypred.round()
```

```
sum = np.dot(inputs,weights) + 0.02
ypred = activation(sum)
error = outputs - ypred
error
```

```
### Create a gradient descent function
def gradient(x,e):
  return e * (x*(1-x))
```

```
sum = np.dot(inputs,weights) + 0.02
ypred = activation(sum)
error = outputs - ypred
chng_weight = gradient(ypred,error)   ### gradient descent
weights = weights + np.dot(inputs.T,chng_weight)
ypred
```

```
for i in range(1000):
  sum = np.dot(inputs,weights) + 0.02
  ypred = activation(sum)
  error = outputs - ypred
  chng_weight = gradient(ypred,error)   ### gradient descent
  weights = weights + np.dot(inputs.T,chng_weight)
```
