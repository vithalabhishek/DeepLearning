```
#### Divide data into features and targets
x = data.iloc[:,:8]    ### features
y = data.iloc[:,8]    ## targets
```

```
x.head()
```

```
y.head()
```

```
#### Split the data into training and testing samples
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size=0.85,random_state=1)
```

```
y.value_counts()
```

```
ytrain.value_counts()
```

```
from sklearn.impute import SimpleImputer
si = SimpleImputer(missing_values=0,strategy="mean")
xtrain = si.fit_transform(xtrain)
xtest = si.fit_transform(xtest)
```

```
xtrain
```

```
import tensorflow.keras as tf
```

```
#### Create the model
model = tf.models.Sequential()
### Add the layers
model.add(tf.layers.Dense(64,input_dim=8,activation='relu'))     ## input and a hidden layer
model.add(tf.layers.Dense(128,activation='relu'))   ## hidden layer
model.add(tf.layers.Dense(256,activation='relu'))   ## hidden layer
model.add(tf.layers.Dense(1,activation='sigmoid')) ## output layer
```

```
### Compile the model
model.compile(loss="binary_crossentropy",optimizer='adam',metrics=['accuracy'])
```

```
## Train the model
model.fit(xtrain,ytrain,epochs=300)
```
